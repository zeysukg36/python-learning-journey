# Python Basics- Advanced Built-in Functions

In this notebook, we will explore highly useful built-in functions that streamline data processing. As an MIS professional, writing efficient code is just as important as writing correct code. We will cover:
* 'enumerate()' -> Tracking index + value effortlessly
* 'zip()' -> Pairing multiple datasets together
* 'sorted()' -> Sorting data without altering the original source
* 'sum()', 'min()', 'max()', -> Quick financial calculations

### 1) 'enumerate()' - Index + Value 
Instead of manually managing a counter variable inside a loop, 'enumerate' tracks the iteration count automatically. This is ideal for generating numbered reports.

```python
project_phases = ["Requirement Analysis", "System Design", "Development", "Testing", "Deployment"]

print("--- Standard Enumeration (Starts at 0) ---")
for index, phase in enumerate(project_phases):
    print(f"Step {index}: {phase}")

print("\n--- Enumeration with a Custom Start (Starts at 1) ---")
# Useful for generating user-friendly lists
for step_num, phase in enumerate(project_phases, start=1):
    print(f"Phase {step_num}: {phase}")
```

### 2) 'zip()' - Pairing Lists 
In data science, we often have related data in separate lists (e.g., a list of employee names and a list of their salaries). 'zip()' combines them row by row.

```python
employees = ["Carry", "David", "Smith"]
salaries = [85000, 92000, 78000]

print('--- Payroll Pairing ---')
for emp_name, salary in zip(employees, salaries):
    print(f'Employee: {emp_name} | Salary: ${salary}')

print('\n--- Note on Unequal Lengths ---')
# zip() stops at the shortest list to prevent index out-of-range errors 
departments = ['IT', 'Finance', 'HR', 'Sales']
budgets = [500000, 3000000] # Only 2 budgets provided

for dept, budget in zip(departments, budgets):
    print(f'Dept: {dept} -> Allocated: ${budget}')
```

### 3) 'sorted()' - Non-destructive Sorting
Unlike the '.sort()' method which permanently changes a list, the 'sorted()' function returns a *new* sorted list, preserving the original data structure (crucial for maintaining raw data integrity).

```python
# Raw customer satisfaction scores (out of 100)
satisfaction_scores = [75, 92, 68, 88, 100, 82]

# Acsending order (Default)
ascending_scores = sorted(satisfaction_scores)

# Descending order (Highest to lowest)
descending_scores = sorted(satisfaction_scores, reverse =True)

print("Raw Data (Preserved) :", satisfaction_scores)
print("Low to High          :", ascending_scores)
print("Top Performers First :", descending_scores)
```

### 4) 'sum()', 'min()', 'max()' - Quick Calculations
These functions eliminate the need for manual 'for' loops when finding totals or extremes in financial data.

```python
quarterly_revenue = [125000, 98000, 142000, 110000]

total_revenue = sum(quarterly_revenue)
lowest_quarter = min(quarterly_revenue)
highest_quarter = max(quarterly_revenue)

print("--- 2026 Revenue Summary ---")
print(f"Total Annual Revenue   :${total_revenue}")
print(f"Weakest Quarter        :${lowest_quarter}")
print(f"Strongest Quarter      :${highest_quarter}")
```

### 5) Mini Exercise - Invoice Summary ('zip' + 'sum')
A practical application combining 'zip' to pair product IDs with costs, and 'sum' to calculate the final invoice total.

```python
product_ids = ["PRD-01", "PRD-02", "PRD-03"]
costs = [1200.50, 450.00, 89.99]

print("--- Generating Invoice ---")
for p_id, cost in zip(product_ids, costs):
    print(f"Item: {p_id} -> ${cost}")

invoice_total = sum(costs)
print("-" * 25)
print(f"Total Amount Due: ${invoice_total}")
```

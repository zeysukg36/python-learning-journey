# Lists in Python

This notebook covers how to work with lists in Python.
Lists are one of the most commonly used data structures in data analysis.

**Topics:**
- Creating lists
- Indexing and slicing
- List methods (append, remove, sort, pop)
- Looping through lists

```python
# 1) Creating Lists

courses = ["Python", "SQL", "Excel", "Tableau"]
grades = [85, 92, 78, 95, 88]
mixed = ["Justin", 20, 3.75, True]

print(courses)
print(grades)
print(mixed)
```

```python
# 2) Indexing and Slicing 

print(courses[0])
print(courses[-1])
print(courses[1:3])
print(grades[:3])

```

```python
# 3) List Methods

courses.append("Power BI")
print(courses)

courses.remove("Excel")
print(courses)

grades.sort()
print(grades)

last = grades.pop()
print(last)
print(grades)
```

```python
# 4) Looping Through Lists

for course in courses:
    print(f"Course: {course}")
    
```

```python
# 5) Mini Practice - Grade Summary

scores = [72, 85, 91, 68, 95, 78, 88]

print(f"Highest score: {max(scores)}")
print(f"Lowest score: {min(scores)}")
print(f"Average score: {sum(scores) / len(scores):.2f}")
print(f"Total score: {len(scores)}")
```

### Mini-Project 1: E-Commerce Cart Evaluation & Discount System 
In this scenario, we evaluate a customer's shopping cart to determine if they are eligible for free shippping or special VIP discounts. This demonstrates the use of **Booleans**, **Comparison Operators**, and basic **If/Else** statements.

```python
# Customer Data
cart_total = 1250.00
is_vip_member = True
customer_points = 450

print("--- Cart Evaluation System ---")

# 1. Checking Data Types (Boolean)
print(f"Data type of 'is_vip_member': {type(is_vip_member)}")

# 2. Applying Logical Operators (and, or)
# Rule: Free shipping if cart is over 1000 OR if the customer is a VIP.
free_shipping_eligible = (cart_total >= 1000) or is_vip_member

print(f"Eligible for free shipping? {free_shipping_eligible}")

# 3. Control Flow (If/Else) with Comparison
# Rule: Apply a 10% discount ONLY IF the cart is over 1500 AND they are a VIP.
if cart_total >= 1500 and is_vip_member:
    discount = cart_total * 0.10
    print(f"Action: VIP Discount applied. You saved {discount} Tl.")
else:
    print(f"Action: Standard pricing applied. No special discount.")

# 4. Short If (Ternary Operator) for quick status assignment
point_status = "High Value Customer" if customer_points > 500 else "Standard Customer"
print(f"Customer Segment: {point_status}")
```

### Mini-Project 3: Loan Approval Decision Engine
Financial institutions rely heavily on algorithms to make quick decisions. This project uses the **if-elif-else** structure to categorize loan application risk based on income and credit score. 

```python
# Applicant Profile
applicant_income = 45000
credit_score = 720

print("\n-- Loan Approval Engine --")

# Multi-condition decision making (if/elif/else)
if credit_score >= 750 and applicant_income >= 50000:
    print("Decision: Automatically Approved (Tier 1- Lowest Interest Rate)")

elif credit_score >= 700:
    print("Decision: Conditionally Approved (Tier 2- Standard Interest Rate)")

elif credit_score >= 650 and applicant_income >= 60000:
    # High income can sometimes compensate for a mediocre credit score
    print("Decision: Manual Review Required by Loan Officer")

else:
    print("Decision: Application Declined (Does not meet minimum risk criteria)")
```

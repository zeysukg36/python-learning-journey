# Python - Clean Code

Clean code is code that is easy to **read, understand, and maintain**.
Writing clean code is a habit - and it matters as much as making code work.

**Topics covered:**
- Meaningful variable names
- Small, single-purpose functions
- DRY principle (Don't Repeat Yourself)
- Avoiding magic numbers
- Basic input validation
- PEP8 style conventions
- Docstrings
- Type hints (optional but useful)

## 1. Meaningful Names (the most critical rule)

Names should tell you **what** something is - without needing a comment to explain it.

```python
# BAD - what is x? what is r? what is t?
x = 1500
r = 0.05
t = x * r
print(t)
```

```python
# GOOD - self-explanatory
monthly_salary = 1500
tax_rate       = 0.05
tax_amount     = monthly_salary * tax_rate
print(tax_amount)
```

## 2. Functions: Small and Single-Purpose

Each function should do **one thing only**.
If a function does too much, split it up.

```python
# BAD - calculates and prints all in one block, hard to reuse
scores = [85, 90, 78, 92]
total  = 0
for s in scores:
    total += s
average = total / len(scores)
print("Total:", total, "Average:", average)
```

```python
# GOOD - logic is reusable and testable
def summarize_scores(scores):
    total   = sum(scores)
    average = total / len(scores)
    return total, average

scores = [85, 90, 78, 92]
total, average = summarize_scores(scores)
print("Total  :", total)
print("Average:", average)
```

## 3. DRY - Don't Repeat Yourself

If you write the same logic more than once, turn it into a function.

```python
# BAD - same formula copy-pasted 3 times
s1 = 500 * 1.08
s2 = 750 * 1.08
s3 = 1200 * 1.08
print(s1, s2, s3)
```

```python
# GOOD - one function, call it as many times as needed
def add_tax(amount, tax_rate=0.08):
    return amount * (1 + tax_rate)

print(add_tax(500))
print(add_tax(750))
print(add_tax(1200))
```

### 4. Avoid Magic Numbers - Name Your Constants

A bare number like '0.08' in the middle of code tells the reader nothing.
Give it a name so the intent is clear.

```python
# BAD - where did 0.08 come from?
salary = 3000
print(salary * 0.08)
```

```python
# GOOD - the constant is named and easy to update
TAX_RATE = 0.08
salary = 3000
print(salary * TAX_RATE)
```

## 5. Basic Input Validation

Always check your inputs before using them.
An unchecked empty list, for example, will crash your program.

```python
# BAD - crashes on empty list
def get_average(numbers):
    return sum(numbers) / len(numbers)

# print(get_average([])) # ZeroDivisionError
```

```python
# GOOD - handles the edge case
def get_average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

print(get_average([10, 20, 30]))
print(get_average([]))
```

## 6. PEP8 - Basic Style Rules

PEP8 is Python's official style guide. Key rules:

- **4 spaces** for indentation
- Variable and function names: 'snake_case'
- Constants: 'UPPER_CASE'
- Keep lines under **79-100 characters**
- One blank line between logical blocks inside a function
- Two blank lines between top-level functions

## 7. Docstrings - Describe What Your Function Does


```python
def get_username(email):
    """
    Extracts the username from an email address.
    Example: 'anna@example.com' -> 'anna'
    """
    return email.split("@")[0]

print(get_username("anna@example.com"))
```

## 8. Type Hints (Optional but Readable)

Type hints are not required, but they make your code easier to understand at a glance.

```python
def calculate_bonus(salary: float, rate: float) -> float:
    return salary * rate

print(calculate_bonus(3000, 0.10))
```

## 9. Mini Example - Putting It All Together

**Goal:** Given a list of products and prices, calculate tax-included totals and overall sum.

```python
TAX_RATE = 0.08

def add_tax(price: float) -> float:
    """Returns the price with tax applied."""
    return price * (1 + TAX_RATE)

def total_cost(prices: list) -> float:
    """Returns the sum of all tax-included prices."""
    return sum(add_tax(p) for p in prices)

products = ["Monitor", "Keyboard", "Mouse"]
prices   = [400, 80, 35]

for product, price in zip(products, prices):
    print(f"{product}: ${add_tax(price):.2f} (tax included)")

print("Total:", round(total_cost(prices), 2))
```

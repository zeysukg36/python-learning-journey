# Python Basics - Loops (for / while)

In this notebook, we will learn how to automate repetitive tasks, which is a core skill in Management Information Systems. We will cover:
* 'for' loops (iterating over lists and ranges)
* 'while' loops (condition-based iteration)
* Control statements: 'break' and 'continue'
* Pratical business exercises

### 1) The 'for' Loop (Iterating over a list)
Used to iterate through a sequence, like a list of corporate assets or employees.

```python
it_assets = ["Server", "Workstation", "Router", "Switch"]

for asset in it_assets:
    print("Deploying update to:", asset)
```

### 2) The 'for' Loop with 'range()'
Generates a sequence of numbers. Useful for running a process a spesific number of times.


```python
# Generating 5 sequential transaction IDs
for i in range(5):
    print("Generated TXN_ID:", i) # Outputs 0, 1, 2, 3, 4
```

### 3) 'range(start, stop, step)'

```python
# Simulating bi-annual financial audits (Every 2 years)
for year in range(2000, 2027, 2):
    print("Running audit for year:", year)
```

### 4) Calculating a Total (Accumulation)
A very common task in business logic is accumulating a total value, such as calculating total monthly expenses.


```python
department_expenses = [1200, 3400, 850, 4100]

total_expense = 0
for expense in department_expenses:
    total_expense += expense

average_expense = total_expense / len(department_expenses)

print("Total Department Expense: $", total_expense)
print("Average Expense per Category: $", average_expense)
```

### 5) Tracking Index + Value manually
Sometimes we need to know the exact position (row number) of the data we are processing.

```python
tasks = ["Data Cleaning", "Model Training", "Deployment"]
index = 0

for task in tasks:
    print(f"Task {index} -> {task}")
    index += 1
```

### 6) The 'while' Loop (Runs as long as a condition is True)
Perfect for system retry mechanisms or processes that must run until a specific state is reached.


```python
retry_attempt = 1

while retry_attempt <= 3:
    print("Connecting to database... Attempt:", retry_attempt)
    retry_attempt += 1
```

### 7) 'break' (Terminates the loop entirely)
Used as an emergency stop. If a critical error or specific condition is met, we halt the loop.

```python
# Processing batches 1 to 9
for batch in range(1, 10):
    if batch == 4:
        print("CRITICAL ERROR at batch 4. Halting all operations!")
        break
    print("Successfully processed batch:", batch)
```

### 8) 'continue' (Skips the current iteration)
Used to skip corrupted or irrelevant data without stopping the entire process.

```python
for user_id in range(1, 6):
    if user_id == 3:
        print("User 3 profile is locked. Skipping...")
        continue
    print("Exporting data for User:", user_id)
```

### 10) Mini Exercise 1- Even Numbers (Filtering Data)
Finding and processing only even-numbered invoices from 1 to 10.

```python
for invoice_num in range(1, 11):
    if invoice_num % 2 == 0:
        print("Processing Even Invoice #", invoice_num)
        
```

### 11) Mini Exercise 2 - Shopping Cart Total
Finding the sum of a list of product prices.

```python
cart_prices = [1299.99, 45.50, 300.00, 19.99]

cart_total = 0
for price in cart_prices:
    cart_total += price

print("Final Checkout Total: $", cart_total)
```

### 12) Mini Exercise 3- Finding the Maximum Value
An algorithm to find the highest sales figure in a dataset without using built-in max() functions.


```python
sales_figures = [15000, 8500, 24000, 12000, 19000]

# Assume the first item is the highest initially
highest_sale = sales_figures[0]

for sale in sales_figures:
    if sale > highest_sale:
        highest_sale = sale # Update if we find a bigger number

print("The best performing sales figure is: $", highest_sale)
```

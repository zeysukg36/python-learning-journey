# Python Sets

A set is an **unordered**, **mutable** collection that only holds **unique** elements.
Duplicates are automatically removed when a set is created.

**What we'll cover:**
- What is a set and how duplicates are handled
- Type checking with 'type()'
- Creating an empty set
- Getting the size with 'len()'
- Adding elements: 'add()'
- Removing elements: 'remove()' vs 'discard()'
- Membership check: 'in'
- Iterating over a set
- Set operations: union, intersection, difference
- Removing duplicates from a list

## 1. What is a Set? (Duplicates are removed automatically)

```python
tags = {"python", "data", "python", "sql"}
print(tags)
```

## 2. Type Check: 'type()'

```python
tags = {"python", "data", "sql"}
print(type(tags)) #class 'set'>
```

## 3. Creating an Empty Set
'{}' creates a **dict**, not a set.
To create an empty set, use 'set()'.

```python
empty_set = set()
empty_dict = {}

print(type(empty_set))   # <class 'set'>
print(type(empty_dict))  # <class 'dict'>
```

## 4. Number of Elements: 'len()'

```python
courses = {"Python", "SQL", "Pandas"}
print(len(courses))
```

## 5. Adding an Element: 'add()'


```python
courses = {"Python", "SQL"}
courses.add("Pandas")
print(courses)
```

## 6. Removing an Element: 'remove()' vs 'discard()'

- 'remove()' raises a **KeyError** if the element doesn't exist
- 'discard()' does nothing if the element doesn't exist

```python
courses = {"Python", "SQL", "Pandas"}

courses.remove("SQL")
print(courses)

# discard won't raise an error
courses.discard("Excel")
print(courses)
```

## 7. Membership Check: 'in'

```python
courses = {"Python", "SQL", "Pandas"}

print("Python" in courses)
print("Excel" in courses)
```

## 8. Iterating Over a Set

Sets are unordered - the iteration order may vary.

```python
courses = {"Python", "SQL", "Pandas"}

for course in courses:
    print("Courses:", course)
```

## 9. Set Operations (Union / Intersection / Difference)

```python
a = {"python", "sql", "pandas"}
b = {"sql", "excel", "power bi"}

print("Union (|)          :", a | b)
print("Intersection (&)   :", a & b)
print("Difference (a-b)   :", a - b)
print("Difference (b-a)   :", b - a)
```

## 10. Removing Duplicates from a List

```python
customers = ["Jason", "Bob", "Jason", "Henry", "Bob"]

unique_customers = set(customers)
print(unique_customers)

# Convert back to list (order not guaranteed)
customer_list = list(unique_customers)
print(customer_list)
```

## 11. Mini Exercise - Finding Common Courses

Two students' courses are given as sets. Find the courses they share.

```python
student1 = {"Python", "SQL", "Pandas"}
student2 = {"SQL", "Excel", "Python"}

common = student1 & student2
print("Common courses:", common)
```

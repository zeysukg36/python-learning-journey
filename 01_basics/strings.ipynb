# Strings in Python

This notebook explores how to work with text data in Python.
Covers string methods, indexing, slicing, and formatting.

**Topics:**
- String basics
- String methods (lower, upper, strip, replace, split)
- Indexing and slicing
- f-string formatting

```python
# 1) String Basics

city = "Istanbul"
product = "Laptop"
message = "Hello, World!"

print(city)
print(product)
print(message)
print(type(city))
```

```python
# 2) String Methods

text = "  management information systems  "

print(text.upper())
print(text.lower())
print(text.strip())
print(text.strip().replace("management","data"))
```

```python
# 3) Indexing and Slicing

word = "DataScience"

print(word[0])     # First character
print(word[-1])    # Last character
print(word[0:4])   # Data
print(word[4:])    # Science
```

```python
# 4) Split Method

course = "Python, SQL, Excel, Tableau"
tools = course.split(", ")
print(tools)
```

```python
# 5) f-string Formatting

name = "Tom"
university = "Izmır"
department = "Management Information Systems"
year = 3

print(f"Name: {name}")
print(f"University: {university}")
print(f"Department: {department}")
print(f"Year: {year}")
```

```python
# 6) Mini Practice - Email Formatter

first_name = "Tom"
last_name = "smith"
domain = "gmail.com"

email = f"{first_name.lower()}.{last_name.lower()}@{domain}"
print(f"Generated email: {email}")
print(f"Username: {email.split('@')[0]}")
print(f"Domain: {email.split('@')[1]}")
```

# Python Functions

A function lets you write a block of code once and reuse it whenever needed. 
Instead of repeating the same logic, you define it once and call it by name.

**Topics covered:**
- What is a function and why use it
- Defining a function with 'def'
- Functions with parameters
- Returning values with 'return'
- Default parameter values
- Using 'if' inside a function
- Returning multiple values
- Mini exercises

### 1. What is a Function?

A function packages repeatable logic into a single named block.
You define it once, call it as many times as needed.

### 2. The Simplest Function

```python
def greet():
    print("Hello!")

greet()
greet()
```

### 3. Function with a Parameter

```python
def welcome(name):
    print(f"Welcome, {name}!")

welcome("Anna")
welcome("Blake")
```

### 4. Returning a Value with 'return'

```python
def add(a, b):
    result = a + b
    return result

total = add(10, 5)
print(total)
```

### 5. Default Parameter Value

If no argument is passed, the default value is used.

```python
def apply_discount(price, discount_rate=0.10):
    return price * (1 - discount_rate)

print(apply_discount(200))       # 10% default
print(apply_discount(200, 0.25)) # 25% custom
```

### 6. Using 'if' Inside a Function


```python
def is_eligible(age):
    if age >= 18:
        return True
    else:
        return False

print(is_eligible(16))
print(is_eligible(21))
    
```

### 7. Returning Multiple Values

A Python function can return more than one value - they come back as a tuple.

```python
def order_summary(prices):
    total   = sum(prices)
    count   = len(prices)
    average = total / count
    return total, count, average

prices = [49, 120, 35, 80]
total, count, average = order_summary(prices)

print("Total    :", total)
print("Items    :", count)
print("Average  :", average)
```

### 8. Mini Exercises

### Exercise 1 - BMI Calculator

Write a function that takes weight (kg) and height (m), and returns the BMI.
Formula: 'BMI = weight / height²'

```python
def calculate_bmi(weight, height):
    return weight / (height ** 2)

print(calculate_bmi(70, 1.75))
```

### Exercise 2 - Domain Extractor

Write a function that takes an email address and returns the domain part.


```python
def get_domain(email):
    return email.split("@")[1]

print(get_domain("anna@gmail.com"))
print(get_domain("blake@company.org"))
```

### Exercise 3 - Highest and Lowest Price

Write a function that takes a list of prices and returns both the max and min values.


```python
def price_range(prices):
    return max(prices), min(prices)

high, low = price_range([49, 120, 35, 80, 200])
print("Highest:", high)
print("Lowest :", low)
```

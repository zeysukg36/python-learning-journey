# Python Basics - Dictionary

In this notebook, we will learn:

- What is a dictionary? Checking with 'type()'
- Key / Value logic
- Reading, adding, updating, deleting values
- 'keys()', 'values()', 'items()'
- Checking keys with the 'in' operator
- Iterating through a dictionary with a loop

## 1) What is a Dictionary?

```python
book = {
    "title": "The Alchemist",
    "author": "Paulo Coelho",
    "pages": 208
}

print(book)
```

## 2) Type check: 'type()'

```python
book = {"title": "The Alchemist",
"author": "Paulo Coelho"}
print(type(book))   # dict 
```

## 3) Reading values (access by key)

```python
book = {"title": "The Alchemist",
"author": "Paulo Coelho", "pages": 208}

print(book["title"])
print(book["pages"])
```

## 4) Safe reading: 'get()'

If the key doesn't exist, it returns 'None' instead of raising an error (or returns your default value).

```python
book = {"title": "The Alchemist",
        "author": "Paulo Coelho"}

print(book.get("title"))
print(book.get("price"))
# None
print(book.get("price", "Not listed"))
# defaul value
```

## 5) Adding and updating

```python
book = {"title": "The Alchemist",
"author": "Paulo Coelho"}

# adding
book["language"] = "Portuguese"

# updating
book["author"] = "Paulo Coelho (translated by Alan R. Clarke)"

print(book)
```

## 6) Deleting: 'pop()' and 'del'

```python
book = {"title": "The Alchemist", "author": "Paulo Coelho", "language": "Portuguese"}

removed = book.pop("language")
print("Removed:", removed)
print(book)

# deleting with del
del book["author"]
print(book)
```

## 7) 'keys()', 'values()', 'items()'

```python
movie = {"title": "Interstellar", "year": 2014, "rating": 8.6}

print(movie.keys())
print(movie.values())
print(movie.items())
```

## 8) Iterating through a dictionary (for loop)

```python
movie = {"title": "Interstellar", "year": 2014, "rating": 8.6}

for key, value in movie.items():
    print(key, "->", value)
```

## 9) Checking keys with 'in'

Use 'in' to check whether a key exists in dictionary.

```python
movie = {"title": "Interstellar", "year": 2014, "rating": 8.6}

print("title" in movie)     # True
print("director" in movie)  # False

if "rating" in movie:
    print("Movie rating:", movie["rating"])
else:
    print("Rating not available.")
```

## 10) Mini exercise - Person card

Create a person dictionary, safely read a missing key with 'get()', then print all info using a loop.

```python
person = {
    "name": "Emma",
    "age": 23,
    "city": "London"
}

# safe read for missing key
print(person.get("job", "Not specified"))

# add new info
person["job"] = "Data Scientist"

# print all info
for key, value in person.items():
    print(f"{key}: {value}")
```

## 1. What is a Class?

A **class** is a blueprint. An **object** is a real instance built from that blueprint.
For example: 'Car' is a class - your specific red Toyota is an object.

## 2. First Class Example - Employee

```python
class Employee:
    def __init__(self, name, age, department):
        self.name       = name
        self.age        = age
        self.department = department

    def display_info(self):
        print(f"Name: {self.name} | Age: {self.age} | Dept: {self.department}")
        

emp1 = Employee("Anna",  28, "Data Science")
emp2 = Employee("Blake", 34, "Engineering")

emp1.display_info()
emp2.display_info()
```

### What does 'self' mean?

'self' refers to the current object.
When you write 'self.name', you are accessing that specific object's data - not another object's.

## 3. Updating an Attribute

```python
class Employee:
    def __init__(self, name, department):
        self.name       = name
        self.department = department


emp = Employee("Anna", "Marketing")
print(emp.name, emp.department)

# Update department
emp.department = "Data Science"
print(emp.name, emp.department)
```

## 4. Class Example - Product

```python
class Product:
    def __init__(self, name, price, stock):
        self.name  = name
        self.price = price
        self.stock = stock

    def apply_discount(self, rate):
        # rate: 0.20 means 20% discount
        self.price = self.price * (1 - rate)

    def in_stock(self):
        return self.stock > 0


item = Product("Headphones", 250, 8)
print(item.name, item.price, item.stock)

item.apply_discount(0.20)
print("Discounted price:", item.price)

print("In stock?", item.in_stock())
```

## 5. Mini Project - Library Book Tracker

A class that tracks whether a book has been borrowed or returned.

```python
class Book:
    def __init__(self, title, author, available=True):
        self.title     = title
        self.author    = author
        self.available = available

    def borrow(self):
        if self.available:
            self.available = False
            print(f"'{self.title}' has been borrowed.")
        else:
            print(f"'{self.title}' is currently unavailable.")

    def return_book(self):
        self.available = True
        print(f"'{self.title}' has been returned. Thank you!")

    def status(self):
        state = "Available" if self.available else "Borrowed"
        print(f"'{self.title}' by {self.author} - {state}")


book = Book("Deep Learning", "Ian Goodfellow")
book.status()

book.borrow()
book.borrow()   # try to borrow again

book.return_book()
book.status()
```

### 6) Multiple object usage

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.balance = balance
    def withdraw(self, amount):
        self.balance -= amount
        print(f"${amount} withdrawn. New balance: ${self.balance}")
    def deposit(self, amount):
        self.balance += amount
        print(f"${amount} deposited. New balance: ${self.balance}")


account1 = BankAccount("Anna", 1000)
account2 = BankAccount("Blake", 200)

account1.withdraw(150)
account2.deposit(500)
```

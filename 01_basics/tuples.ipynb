# Python Basics - Tuples
In this notebook, we will learn:
* What is a tuple? Checking with 'type()'
* Tuple vs List (Key difference: Tuples are **immutable**
* Accessing elements (indexing) and slicing
* Finding length with 'len()'
* Tuple unpacking and multiple assignments
* 'count()' and 'index()' methods


### 1) What is a Tuple?
Tuples are used to store multiple items in a single variable. They are written with round brackets '()'.

```python
departments = ("HR", "Finance", "IT")
employee_ratings = (4.5, 3.8, 4.9)

print(departments)
print(employee_ratings)
```

### 2) Type Check: 'type()'

```python
departments = ("HR", "Finance", "IT")
print(type(departments)) # Should return <class 'tuple'>
```

### 3) Single-element Tuple (Attention!)
To create a tuple with only one item, you must add a comma after the item. Otherwise, Python will not recognize it as a tuple.

```python
single_tuple = ("Python",)
not_a_tuple = ("Python")

print(type(single_tuple))   # <class 'tuple'>
print(type(not_a_tuple))    # <class 'str'>
```

### 4) Accessing Elements (Index)

```python
departments = ("HR", "Finance", "IT", "Marketing")

print(departments[0])    # First element
print(departments[-1])   # Last element
```

### 5) Slicing


```python
departments = ("HR", "Finance", "IT", "Marketing", "Sales")

print(departments[0:3])    # First 3 departments
print(departments[2:])     # From index 2 to the end    
```

### 6) Length: 'len()'


```python
departments = ("HR", "Finance", "IT")
print(len(departments))
```

### 7) Tuples are Immutable
Once a tuple is created, you cannot change its values. Tuples are unchangeable.

```python
departments = ("HR", "Finance", "IT")

# The following line would raise a TyppeError if uncommented
# departments[0] = "Security"

print(departments)
```

### 8) Tuple Unpacking
Extracting the values back into separate variables is called "unpacking".

```python
product_data = ("MacBook Pro", 2500, 15) # (name, price, stock)

prod_name, price, stock = product_data

print("Product:", prod_name)
print("Price: $", price)
print("Stock Quantity:", stock)
```

### 9) Swapping Variables via Multiple Assignment


```python
usd_rate = 32.5
eur_rate = 35.1

# Swapping the values
usd_rate, eur_rate = eur_rate, usd_rate

print("USD Rate:", usd_rate)
print("EUR Rate:", eur_rate)
```

### 10) 'count()' and 'index()'

```python
status_codes = (200, 404, 200, 500, 200)

print("How many 200 codes? :", status_codes.count(200))
print("Index of 500 code?  :", status_codes.index(500))
```

# Python Error Handling - Try / Except

Programs crash when they encounter unexpected input or situations.
'try / except' lets us handle these errors gracefully instead of crashing.

**Topics covered:**
- What is an exception?
- Basic 'try / except'
- Catching specific errors: 'ValueError', ' ZeroDivisionError', 'FileNotFoundError'
- Catching multiple errors
- Seeing the error message with 'as e'
- 'else' block (runs if no error)
- 'finally' block (always runs)
- Real-world mini exercises

### 1. What Happens Without Error Handling?

The code below will crash - it tries to convert a word into a number.


```python
# This will raise an error
age = int("twenty")
print(age)
```

## 2. Basic 'try / except'

```python
try:
    age = int("twenty")
    print("Age:", age)
except:
    print("Error: could not convert to a number")
```

### 3. Catching a Specific Error (ValueError)

```python
value = "abc"

try:
    number = int(value)
    print("Number:", number)
except ValueError:
    print(f"ValueError: '{value}' cannot be converted to an integer")
    
```

### 4. Catching Multiple Errors

Handle 'ValueError' and 'ZeroDivisionError' separately.

```python
def divide(a, b):
    return a / b

try:
    x = int("10")
    y = int("0")
    print(divide(x, y))
except ValueError:
    print("ValueError: invalid number format")
except ZeroDivisionError:
    print("ZeroDivisionError: cannot divide by zero")
```

### 5. Seeing the Error Message ('except ... as e')


```python
try:
    result = int("hello")
except ValueError as e:
    print("Error message:", e)
```

## 6. 'else' Block (runs only if no error occurred)


```python
raw = "42"

try:
    number = int(raw)
except ValueError:
    print("Conversion failed")
else:
    print("Successful conversion:", number)
```

## 7. 'finally' Block (always runs)

```python
try:
    number = int("xyz")
except ValueError:
    print("Error caught")
finally:
    print("finally: this line always executes")  
```

## 8. File Reading Example (FileNotFoundError)

```python
filename = "missing_file.txt"

try:
    with open(filename, "r", encoding="utf-8") as f:
        content = f.read()
        print(content)
except FileNotFoundError:
    print(f"FileNotFoundError: '{filename}' not found")
```

## 9. Real-World Example - User Input Validation

Try running this cell and entering both a valid number and a word.

```python
try: 
    # user_input = input("Enter a number: ")
    print("Simulation: User entered '10'")
    user_input = "10" 

    
    number = int(user_input)
    print("Double:", number * 2)
except ValueError:
    print("Error: please enter a valid number")
```

## 10. Mini Exercise - Safe Division Function

Write a function that divides two numbers.
If division by zero occurs, return a message instead of crashing.

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: division by zero"

print(safe_divide(10, 4))
print(safe_divide(10, 0))
```

## 11. Mini Exercise - List Index Error

Try to access an index that doesn't exist.
Catch the error and print a friendly message.

```python
products = ["Laptop", "Mouse", "Keyboard"]

try:
    print(products[10])
except IndexError:
    print("IndexError: that index does not exist in the list")
```

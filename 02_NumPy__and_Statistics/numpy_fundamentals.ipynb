## 1. Setup: Import NumPy and Set Seed


```python
import numpy as np

np.random.seed(7)  # ensures reproducible results
print("NumPy version:", np.__version__)
```

## 2. What is NumPy? The `ndarray` Structure

NumPy is a high-performance library for numerical computing.  
Its core data structure is `ndarray` — optimized for vector and matrix operations.

```python
py_list = [10, 20, 30, 40]
arr     = np.array(py_list)

print("List:", py_list, type(py_list))
print("Array:", arr,   type(arr))
```

### 2.1 Core Properties: `shape`, `dtype`, `ndim`, `size`


```python
arr = np.array([15, 30, 45, 60, 75])

print("array :", arr)
print("shape :", arr.shape)
print("dtype :", arr.dtype)
print("ndim  :", arr.ndim)
print("size  :", arr.size)

```

## 3. Array Creation Methods



```python
zeros    = np.zeros((2, 4))
ones     = np.ones((3, 3))
rang     = np.arange(0, 50, 5)        # 0 to 50, step 5
spaced   = np.linspace(0, 1, 6)       # 6 evenly spaced points between 0 and 1
identity = np.eye(4)                  # 4x4 identity matrix

print("Zeros:\n",    zeros)
print("\nOnes:\n",   ones)
print("\nRange:",    rang)
print("\nLinspace:", spaced)
print("\nIdentity:\n", identity)
```

### 3.1 Random Arrays (Foundation of Synthetic Data)

```python
# Standard normal distribution (mean=0, std=1)
rand_normal  = np.random.randn(5)

# Uniform distribution between 0 and 1
rand_uniform = np.random.rand(5)

# Random integers between 18 and 65
rand_int     = np.random.randint(18, 65, size=8)

print(f"Normal  : {np.round(rand_normal, 2)}")
print(f"Uniform : {np.round(rand_uniform, 2)}")
print(f"Int     : {rand_int}")
```

## 4. 2D Arrays — Matrix Creation and Access

```python
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print(matrix)
print("Shape:", matrix.shape)
```

### 4.1 Indexing and Slicing

```python
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Row 0          =>", matrix[0])
print("Element [1,2]  =>", matrix[1, 2])
print("Column 1       =>", matrix[:, 1])
print("Top-left 2x2:\n",   matrix[:2, :2])
```

## 5. Boolean Masking and Filtering

```python
scores = np.array([45, 72, 88, 34, 91, 60, 55, 79])
mask   = scores >= 70

print("Mask          :", mask)
print("Scores >= 70  :", scores[mask])
```

### 5.1 Multiple Conditions (and / or)

Use `&` for AND, `|` for OR. Always wrap conditions in parentheses.

```python
scores = np.array([45, 72, 88, 34, 91, 60, 55, 79])
mask   = (scores >= 60) & (scores <= 85)

print("Scores between 60 and 85:", scores[mask])
```

## 6. Vectorization: Why NumPy is So Fast

Python loops process elements one by one.  
NumPy vectorizes the same operation across the entire array in a single step.

```python
revenue  = np.array([500, 1200, 340, 870, 2100])
MARGIN   = 0.30
profit   = revenue * MARGIN        # no loop needed

print("Revenue :", revenue)
print("Profit  :", profit)
```

### 6.1 Broadcasting (Shape Alignment)

```python
monthly_sales = np.array([
    [1000, 1500, 1200],
    [900,  1100, 1300]
])

bonus = np.array([100, 200, 150])   # 1x3 — broadcasts across both rows
adjusted = monthly_sales + bonus

print("Original:\n",  monthly_sales)
print("Adjusted:\n",  adjusted)
```

## 7. Aggregation and `axis`

```python
matrix = np.array([
    [5, 8, 2],
    [3, 7, 9]
])

print("Total (all elements)    :", matrix.sum())
print("Row totals   (axis=1)   :", matrix.sum(axis=1))
print("Column totals (axis=0)  :", matrix.sum(axis=0))
print("Mean                    :", matrix.mean())
print("Std                     :", matrix.std())
```

## 8. `reshape`, `ravel`, `transpose`

```python
arr    = np.arange(1, 13)          # 1 to 12
matrix = arr.reshape(3, 4)

print("Original:", arr)
print("Matrix (3x4):\n", matrix)

flat = matrix.ravel()
print("Flattened:", flat)

print("Transposed:\n", matrix.T)
```

## 9. Synthetic Dataset (Mini Scenario)

We'll generate a fictional employee dataset with 5 features — fully synthetic.

- `age` (22–60)
- `salary` (normal distribution)
- `experience` (tied to salary with noise)
- `performance_score` (0–100)
- `tenure` (months, exponential distribution)
- `target` (high performer? — simple rule)

```python
N = 10_000  # 10,000 employees

age        = np.random.randint(22, 60, size=N)

salary     = np.random.normal(loc=55000, scale=12000, size=N)
salary     = np.clip(salary, 25000, 120000)

experience = (salary * 0.0003 + np.random.normal(0, 2, size=N)).clip(0, None)

performance = np.random.randint(0, 101, size=N)

tenure      = np.random.exponential(scale=36, size=N).astype(int)  # months

target = (
    (performance >= 75).astype(int) +
    (salary >= 60000).astype(int) +
    (tenure >= 24).astype(int)
)
target = (target >= 2).astype(int)

print("Sample ages       :", age[:5])
print("Sample salaries   :", np.round(salary[:5], 0))
print("Sample performance:", performance[:5])
print("High performer %  :", round(target.mean() * 100, 1))
```

### 9.1 Stack All Features into One Matrix

```python
X = np.column_stack((age, salary, experience, performance, tenure))

print("X shape     :", X.shape)
print("First row   :", X[0])
```

### 9.2 Quick Summary Statistics


```python
print("Age         — mean:", round(age.mean(), 1),        "| std:", round(age.std(), 1))
print("Salary      — mean:", round(salary.mean(), 0),     "| std:", round(salary.std(), 0))
print("Experience  — mean:", round(experience.mean(), 1), "| std:", round(experience.std(), 1))
print("Performance — mean:", round(performance.mean(), 1))
```

### 9.3 Filter Example: Senior high earners

Select employees older than 40 with salary above 70,000.


```python
mask     = (age > 40) & (salary > 70000)
selected = salary[mask]

print("Matching employees :", selected.size)
print("Avg salary (group) :", round(selected.mean(), 0))
```

## 10. Normalization and Standardization

Two common preprocessing steps before machine learning:

- **Min-Max Normalization** → scales values to [0, 1]
- **Z-score Standardization** → centers around mean=0, std=1

```python
# Min-Max Normalization
salary_minmax = (salary - salary.min()) / (salary.max() - salary.min())
print("Min-Max (first 5):", np.round(salary_minmax[:5], 3))

# Z-score Standardization
salary_z = (salary - salary.mean()) / salary.std()
print("Z-score (first 5):", np.round(salary_z[:5], 3))
```

## 11. Performance Test: Python List vs NumPy Array

Same operation, two approaches — let's measure the difference.


```python
import time

N     = 2_000_000
lst   = list(range(N))
arr   = np.arange(N)

# Python loop
t0 = time.perf_counter()
result_list = [x * 3 for x in lst]
t1 = time.perf_counter()

# NumPy vectorized
t2 = time.perf_counter()
result_np = arr * 3
t3 = time.perf_counter()

print(f"Python loop : {round(t1 - t0, 4)} sec")
print(f"NumPy vector: {round(t3 - t2, 4)} sec")
```

### 11.1 Large Matrix Addition


```python
import time

A = np.random.rand(2000, 2000)
B = np.random.rand(2000, 2000)

t0 = time.perf_counter()
C  = A + B
t1 = time.perf_counter()

print(f"2000x2000 matrix addition: {round(t1 - t0, 4)} sec")
print("C shape:", C.shape)
```

## 12. Outlier Detection with Z-score

Z-score tells us how many standard deviations a value is from the mean.  
Values with `|z| > 2.5` are typically flagged as outliers.

```python
data = np.array([22, 25, 24, 23, 21, 500, 26, 24, 22, 23], dtype=float)

mu    = data.mean()
sigma = data.std()
z     = (data - mu) / sigma

mask     = np.abs(z) < 2.5
filtered = data[mask]
outliers = data[~mask]

print("Data          :", data)
print("Mean / Std    :", round(mu, 2), "/", round(sigma, 2))
print("Z-scores      :", np.round(z, 2))
print("Filtered      :", filtered)
print("Outliers      :", outliers)
```

# Variables & Numbers in Python

This notebook covers the basics of variables, numeric data types, arithmetic operations, and type conversion in Python. 

**Topics:**
- Variable assignment
- int and float types
- Arithmetic operations
- Type conversion (casting)
- f-string formatting

```python
# 1) Variable Assignment

student_name = "Justin"
student_age = 20
gpa = 3.75
is_enrolled = True 

print(student_name, student_age, gpa, is_enrolled)
```

```python
# 2) Checking Data Types

print(type(student_name))
print(type(student_age))
print(type(gpa))
print(type(is_enrolled))
```

```python
# 3) Arithmetic Operations
# Shopping scenario

price = 250
quantity = 3

total = price * quantity
discount = total * 0.10
final_price = total - discount

print("Total:", total)
print("Discount (10%):", discount)
print("Final Price:", final_price)
```

```python
# 4) Type Conversion (Casting)

str_number = "42"
str_price = "19.99"

converted_int = int(str_number)
converted_float = float(str_price)

print(converted_int, type(converted_int))
print(converted_float, type(converted_float))
```

```python
# 5) f-string Formatting

name = "Justin"
age = 20
gpa = 3.75

print(f"Student: {name}")
print(f"Age: {age}")
print(f"GPA: {gpa:.2f}")
```

```python
# 6) Mini Practice - Student Grade Calculator 

student = "Justin"
midterm = 78
final_exam = 85
homework = 92

average = (midterm * 0.30) + (final_exam * 0.40) + (homework * 0.30)

print(f"Student: {student}")
print(f"Midterm (30%): {midterm}")
print(f"Final Exam (40%): {final_exam}")
print(f"Homework (30%): {homework}")
print(f"Weighted Average: {average:.2f}")
```
